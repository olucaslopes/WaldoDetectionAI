```python
import os

import cv2
from ultralytics import YOLO


class WaldoDetector:
    def __init__(self, model_path, screen_size=(640, 480)):
        self.home_path = os.path.join(os.getcwd().split("WaldoDetectionAI")[0], "WaldoDetectionAI")
        self.model = YOLO(f'{self.home_path}/runs/detect/train/weights/{model_path}')

    def start_video(self, camera_index=0):
        cap = cv2.VideoCapture(camera_index)

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            frame = self.detect(frame)

            cv2.imshow('Video', frame)
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

        cap.release()
        cv2.destroyAllWindows()

    def detect(self, frame):
        return self.model.predict(frame)[0].plot()

def main():
    detector = WaldoDetector('best.pt')
    detector.start_video()

main()
```
